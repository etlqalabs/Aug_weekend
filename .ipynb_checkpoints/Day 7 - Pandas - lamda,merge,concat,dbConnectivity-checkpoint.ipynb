```python
# Merge ( analogy SQL Join )
```

```python
df_emp = pd.read_csv('employees.csv')
```

```python
df_sal = pd.read_csv('salaries.csv')
```

```python
df_emp
```

```python
df_sal
```

```python
# inner join

df_emp.merge(df_sal,on='eno',how='inner')
```

```python
# left outer join

df_emp.merge(df_sal,on='eno',how='left')
```

```python
# right outer join

df_emp.merge(df_sal,on='eno',how='right')
```

```python
#  outer join

df_emp.merge(df_sal,on='eno',how='outer')
```

```python
merged_df = pd.merge(df_emp, df_sal, on='eno', suffixes=('_emp', '_sal'))
```

```python
merged_df
```

```python
merged_df = pd.merge(df_emp, df_sal, left_on='eno', right_on='eno_sal',how='right',suffixes=('_emp', '_sal'))
```

```python
merged_df
```
